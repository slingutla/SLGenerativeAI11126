```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
my_api_key[:5], my_api_key[-10:]
client = OpenAI(api_key=my_api_key)
```

```python
def get_response_from_model(prompt):
    response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=[
        {"role": "system", "content": "You are a helpful assistant. Answer as a csv fields title, description, time_data_extracted"},
        {"role": "user", "content": prompt}
        # {"role": "user", "content": f"Here is the context: \n\ {context} \n\. \n\n Now, answer the question: {prompt}")
    ]
)
    # print (response)
    # print (response.choices[0].message.content)
    return response.choices[0].message.content
```

```python
import time
while True:
    # Ask user for a question
    user_prompt = input("Ask something: ")

    if (user_prompt.lower() != 'quit'):
        # Get and print the response
        response = get_response_from_model(user_prompt)
        print("\nOpenAI says:", response)
        
        # add delay of 3 seconds
        time.sleep(3)
    else:
        print("Exiting the loop. Goodbye!")
        break    
```
