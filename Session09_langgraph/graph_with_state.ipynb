```python
from typing import TypedDict, Optional
from langgraph.graph import StateGraph, START, END

# class MyState extends TypedDict
class MyState(TypedDict):
    # name: Optional[str]
    # age: Optional[int]
    name: str
    age: int
    grade: int
# grade is age - 5
# MyState node_a(MyState state) in Java style
def node_a(state: MyState) -> MyState:
    print("A ran")
    print(state) # state =  {"name": "", "age": 0}
    state["name"] = "Node A"
    state["age"] = 10
    state["grade"] = state["age"] - 5 
    print(f"node_a: {state["name"]} is {state["age"]} years old and grade {state["grade"]}" ) # node_a: Node A is 10 years old
    return state # state = {"name": "Node A", "age": 10}

def node_b(state: MyState) -> MyState:
    print("B ran")
    print(state) # state = {"name": "Node A", "age": 10}
    state["name"] = "Node B"
    state["age"] = (state["age"] or 0) + 5
    state["grade"] = state["age"] - 5
    print(f"node_b: {state["name"]} is {state["age"]} years old and grade {state["grade"]}" ) # node_b: Node B is 15 years old
    return state


builder = StateGraph(MyState)
builder.add_node("A", node_a)
builder.add_node("B", node_b)

builder.add_edge(START, "A")
builder.add_edge("A", "B")
builder.add_edge("B", END)

graph = builder.compile()
```

```python

print(graph.get_graph().draw_ascii())


# initial_state={"name":"START NODE", "age":0}

initial_state={}
result = graph.invoke(initial_state)
print(result)

# Initial state: 
# state = {name:"START NODE", age:0}

# After Node A executes:
# state = {"name": "Node A", "age": 10}


# After Node B executes:
# state = {"name": "Node B", "age": 15}

# Final state after graph execution: 
# state = {"name": "Node B", "age": 15}
```
