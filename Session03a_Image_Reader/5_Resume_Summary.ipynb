```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=my_api_key)
print(f"my api key is:[{my_api_key[:4]}...{my_api_key[-4:]}]")
```

```python
import pdfplumber
pdf_path = "./data/Jyotshna_resume.pdf"

def extract_text_from_pdf(pdf_path):

    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            t = page.extract_text()
            if t:
                text += t + "\n"
    return text.strip()

pdf_text = extract_text_from_pdf(pdf_path)
pdf_text
```

```python
response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=[
        {"role": "system", "content": "Extract Candidate name,Years of experience,Skill set,Education summary "
        "and Then, generate a concise professional summary in JSON format"},
        {"role": "user", "content": pdf_text}
    ]
)

print("Extracted Info from PDF:\n")
print(response.choices[0].message.content)
```

```python
import ollama

response = ollama.chat (
    model="llama3",
    messages=[
        {"role": "system", "content": "Extract Candidate name,Years of experience,Skill set,Education summary "
        "and Then, generate a concise professional summary in JSON format"},
        {"role": "user", "content": pdf_text}
    ]
)

print("Extracted Info from PDF:\n")
print(response["message"]["content"])

```
