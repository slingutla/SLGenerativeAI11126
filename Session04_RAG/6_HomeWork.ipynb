```python
#Write a function that reads a text file and returns its length, handling the case where the file doesnâ€™t exist.

def file_length(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            return len(content)
    except FileNotFoundError:
        return "File does not exist."
```

```python
filepath ="data/ContentLength.txt"
print(f"Length of the file '{filepath}': {file_length(filepath)}")
```
